## Initialising and invoking an agent

```python
from langchain.agents import create_agent

agent = create_agent(model=model)
```

```python
agent = create_agent(model="claude-sonnet-4-5")
```

```python
agent = create_agent("gpt-5-nano")
```

```python
from langchain.messages import HumanMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?")]}
)
```

```python
from pprint import pprint

pprint(response)
```

```python
print(response['messages'][-1].content)
```

```python
from langchain.messages import AIMessage

response = agent.invoke(
    {"messages": [HumanMessage(content="What's the capital of the Moon?"),
    AIMessage(content="The capital of the Moon is Luna City."),
    HumanMessage(content="Interesting, tell me more about Luna City")]}
)

pprint(response)
```

## Streaming Output

```python
for token, metadata in agent.stream(
    {"messages": [HumanMessage(content="Tell me all about Luna City, the capital of the Moon")]},
    stream_mode="messages"
):

    # token is a message chunk with token content
    # metadata contains which node produced the token
    
    if token.content:  # Check if there's actual content
        print(token.content, end="", flush=True)  # Print token
```

Direct Chat Model (First Example)
pythonmodel = init_chat_model(model="gpt-5-nano")
response = model.invoke("What's the capital of the Moon?")

## Answer
Single call to the LLM
Model responds immediately with whatever it generates
No tool use, no reasoning loop
You get one response and done
Simple question → answer pattern

Agent (Second Example)
pythonagent = create_agent("gpt-5-nano")
response = agent.invoke({"messages": [...]})

Agentic loop - the LLM can reason, plan, and act
Can make multiple LLM calls as it thinks through the problem
Can use tools (web search, calculators, databases, etc.)
Can break down complex tasks into steps
Keeps iterating until it reaches a final answer

Practical Example
Direct chat - "What's the capital of the Moon?"
→ Model hallucinates an answer or says "The Moon has no capital"
Agent - "What's the capital of the Moon?"
→ Agent might:

Recognize it needs to search for info
Use a web search tool
Find that the Moon has no capital
Return a factual answer based on the search

Agents are for tasks requiring reasoning, tool use, or multi-step problem solving. Direct chat models are for simpler single-turn conversations where you just need the model's direct response.
